#Engine Interface

```python
class Engine:
    def __init__(self):
        self.speed = 0

    def increase(self):
        self.speed += 1

    def decrease(self):
        self.speed -= 1

    def update_speed(self, car_speed):
        self.speed = car_speed
```

```python
class GasolineEngine(Engine):
    def get_type(self):
        return "Gasoline"
```

```python
class ElectronicEngine(Engine):
    def get_type(self):
        return "Electric"
```

```python
class MixedHybridEngine(Engine):
    def get_type(self):
        return "Hybrid"

    def update_speed(self, car_speed):
        self.speed = car_speed

        if car_speed < 50:
            print("Electric mode")
        else:
            print("Gas mode")
```

```python
class Car:
    def __init__(self, engine):
        self.engine = engine
        self.speed = 0

    def start(self):
        self.speed = 0
        print("Car started")

    def stop(self):
        self.speed = 0
        print("Car stopped")

    def accelerate(self):
        if self.speed < 200:
            self.speed += 20
            self.engine.update_speed(self.speed)


    def brake(self):
        if self.speed > 0:
            self.speed -= 20
            self.engine.update_speed(self.speed)
            print('brake',self.speed)


```

```python
class CarFactory:
    def create_car(engine_type):
        if engine_type == "gas":
            return Car(GasolineEngine())
        elif engine_type == "electric":
            return Car(ElectronicEngine())
        elif engine_type == "hybrid":
            return Car(MixedHybridEngine())
```

```python
car = CarFactory.create_car("hybrid")

car.start()

car.accelerate()  # 20
car.accelerate()  # 40
car.accelerate()  # 60

car.brake()  # 40

car.stop()
```
