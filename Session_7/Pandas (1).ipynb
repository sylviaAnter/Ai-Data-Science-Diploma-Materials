#Missing Value

```python
data = {
    "index": [1, 2, 3, 4, 5],
    "name": ["noha", None, "monica",None, "shady"],
    "age": [45,None, 25, 39, None],
    "job": [None, "data scientist", "data engineer", "data analyst", None]
}

df = pd.DataFrame(data)
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# 👉 معناها:
# احذف أي صف (row) فيه قيمة مفقودة
# (سواء None أو NaN)
df.dropna()
```

```python
df.dropna(axis=1)
```

#deal with NAN

```python
df['name']=df['name'].fillna(value='Not Entered')
df
```

```python
df['job']=df['job'].fillna(value='without job')
df
```

```python
df['age'].fillna(value=df['age'].mean())
```
