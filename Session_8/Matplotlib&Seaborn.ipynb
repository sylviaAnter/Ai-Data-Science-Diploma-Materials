#Scatter
#### rasm bayanii bywd7 el 3alka biin 2 features

```python
age = [15,25,35,45,55]
salaries = [2000,4000,6000,8000,10000]

plt.scatter(age,salaries,color='r')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Relation between Age & salary')
plt.show()
```

#Bar

```python
x=np.arange(10)
y=np.random.randint(1,10,10)
plt.bar(x,y)
plt.show()
```

#create multiplots on same canvas

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,y,'g*-')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'b+-')
plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
plt.show()
```

#Legends

```python
plt.plot(x,x**2,label='First Line')
plt.plot(x,y**2,label='Second Line')
plt.legend()
plt.show()
```

#Seaborn

```python
df=pd.read_excel("/content/Employees.xlsx")
df.head()
```

```python
df.columns
```

#Histogram

```python
sns.histplot(df['Monthly Salary'])
plt.show()
```

#Jointplot
####Jointplot = Scatter + Histogram + Histogram

```python
sns.jointplot(x='Monthly Salary',y='Overtime Hours' ,data=df,kind='scatter')
plt.show()
```

#Heatmap

```python
pivot = df.pivot_table(
    index='Country',            # الصفوف
    columns='Years',        # الأعمدة (أو أي فئة)
    values='Monthly Salary',  # القيم
    aggfunc='mean'            # أو sum, max,
)
```

```python
pivot
```

```python
sns.heatmap(pivot,annot=True, fmt=".0f", cmap="YlGnBu")
plt.show()
```
